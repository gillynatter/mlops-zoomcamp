#### What Is `DictVectorizer`?

`DictVectorizer` is from:

```python
from sklearn.feature_extraction import DictVectorizer
```

It converts **a list of Python dictionaries** into a **numerical feature matrix**.

It is mainly used for:

* Categorical encoding
* One-hot encoding
* Production ML pipelines
* Sparse feature generation
---

#### Why It’s Used in MLOps Zoomcamp

Because it:

* Automatically one-hot encodes categorical features
* Returns a **sparse matrix** (memory efficient)
* Keeps a stable feature mapping between train and validation
* Is easy to serialize with the model
---

#### Step 1: What Your DataFrame Looks Like

Example DataFrame:

```python
import pandas as pd

df = pd.DataFrame({
    "PUlocationID": ["43", "74", "132"],
    "DOlocationID": ["151", "43", "75"]
})

print(df)
```

Output:

```
  PUlocationID DOlocationID
0           43          151
1           74           43
2          132           75
```

---

#### Step 2: Convert DataFrame to Dictionary Format

This is the **exact required format**:

```python
dicts = df.to_dict(orient="records")
print(dicts)
```

Output:

```
[
 {'PUlocationID': '43', 'DOlocationID': '151'},
 {'PUlocationID': '74', 'DOlocationID': '43'},
 {'PUlocationID': '132', 'DOlocationID': '75'}
]
```

Important:

* It becomes a **list**
* Each row becomes a **dictionary**
* Keys = column names
* Values = category values (must be strings)

This format is what `DictVectorizer` expects.
---

#### Step 3: Apply `DictVectorizer`

```python
dv = DictVectorizer()

X = dv.fit_transform(dicts)
```

Now `X` is a **sparse matrix**.
---

#### What Features Look Like Internally

Notice:
Each unique value becomes:
```

```
So:
* `PUlocationID=43`

#### Why Convert to String First?
In the taxi homework:
```python
df[categorical] = df[categorical].fillna(-1).astype(int).astype(str)
```
Why?
Because:
* IDs are categorical, not numeric quantities
* `DictVectorizer` treats numbers differently
* Converting to string forces categorical treatment

#### Train vs Validation Behavior

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
X_val = dv.transform(val_dicts)
```

Important:

* `fit` learns feature mapping
* `transform` uses the same mapping
* Unseen categories are ignored safely

This is why it's production-safe.

#### How Ignored Categories are Treated At Prediction Time 

Perfect — let’s look at how `DictVectorizer` handles **unseen categories**, which is a very important point for production ML.
---

#### Scenario

Suppose you have **training data** with only some location IDs:

```python
from sklearn.feature_extraction import DictVectorizer

# Training data
train_dicts = [
    {'PUlocationID': '43', 'DOlocationID': '151'},
    {'PUlocationID': '74', 'DOlocationID': '43'},
]

dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)

print("Feature names:", dv.feature_names_)
```

**Output:**

```
Feature names: ['DOlocationID=43', 'DOlocationID=151', 'PUlocationID=43', 'PUlocationID=74']
```

So `DictVectorizer` only learned features present in the **training set**.

---

### Validation / New Data with Unseen Category

Suppose your **validation data** has a new category:

```python
val_dicts = [
    {'PUlocationID': '132', 'DOlocationID': '151'},  # PU=132 unseen
    {'PUlocationID': '43', 'DOlocationID': '999'},   # DO=999 unseen
]

X_val = dv.transform(val_dicts)
print(X_val.toarray())
```

**Output (dense for clarity):**

```
[[0 1 0 0]  # PU=132 is unseen → ignored
 [0 0 1 0]]  # DO=999 is unseen → ignored
```

### ✅ Key Points:

1. **Unseen categories are ignored**

   * `PUlocationID=132` or `DOlocationID=999` do not appear in training features → they are silently dropped.
2. **Matrix shape is consistent**

   * `X_val.shape[1] == X_train.shape[1]`
3. **No errors**

   * Unlike `pd.get_dummies()` without column alignment, which can break pipelines.

---

# Why This Matters

* In production, your model may see **new pickup/dropoff locations**.
* Using `DictVectorizer` ensures:

  * Safe handling of unseen categories
  * Fixed feature order
  * Sparse memory efficiency

---
